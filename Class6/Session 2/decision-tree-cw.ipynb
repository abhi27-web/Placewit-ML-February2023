```python
df = pd.read_csv('/kaggle/input/heart-disease-prediction-using-logistic-regression/framingham.csv')
```

```python
from sklearn import tree
```

```python
df.dropna(subset=['cigsPerDay'],inplace=True)
df.dropna(subset=['BMI'],inplace=True)
df.dropna(subset=['heartRate'],inplace=True)
df['totChol']=df['totChol'].replace(np.nan,df['totChol'].mean())
df['BPMeds']=df['BPMeds'].replace(np.nan,df['BPMeds'].mean())
df['education']=df['education'].replace(np.nan,df['education'].mean())
df['glucose']=df['glucose'].replace(np.nan,df['glucose'].mean())
```

```python
df.head()
```

```python
X = df.drop('TenYearCHD', axis = 1)
y = df['TenYearCHD']
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X,y)
```

```python
clf.score(X,y)
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {'criterion':('gini', 'entropy'), 'splitter':['best','random']}
```

```python
dt = tree.DecisionTreeClassifier()
gridsearch = GridSearchCV(dt, parameters)
```

```python
gridsearch.fit(X,y)
```

```python
sorted(gridsearch.cv_results_.keys())
```
